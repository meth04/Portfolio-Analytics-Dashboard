```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('../data/processed/sp500_price.csv', index_col='Date')
data.head()
```

#### Daily return

```python
results = data.pct_change().dropna(how='all')
results.head(5)
```

#### cumulative return

```python
cumret = (1 + results).cumprod() - 1
cumret.head(5)
```

#### Annualized return 

```python
total_days = results.shape[0]
total_days
```

```python
cumulative_return = cumret.iloc[-1]
cumulative_return
```

```python
annualized_return = (1 + cumulative_return) ** (251 / total_days) - 1
print(f"Annualized return: \n{round(annualized_return * 100, 2)}%")
```

#### Annualized Volatility

Annualized Volatility = Daily Volatility x $\sqrt{251}$

where:
+ Daily Volatility is the standard deviation of daily return 
+ 251 is the assumed number of trading days in a year

```python
daily_volatility = results.std()
annualized_volatility = daily_volatility * np.sqrt(251)

print(f'Annualized Volatility: \n{round(annualized_volatility * 100, 2)}%')
```

#### Sharpe Ratio
Sharpe Ratio = (Annualized Return - Risk-Free Rate) / Annualized Volatility

where:

+ Annualized return: the annualized return of the portfolio
+ Risk-free rate: the return of a risk free asset. This is typically a small percentage, such as 0.02

```python
risk_free_rate = 0.02
sharpe_ratio = (annualized_return - risk_free_rate) / annualized_volatility

print(f"Sharpe Ratio: \n{round(sharpe_ratio, 2)}")
```
