```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('../data/processed/sp500_price.csv', index_col='Date')
data.head()
```

#### Daily return

```python
results = data.pct_change().dropna(how='all')
results.head(5)
```

#### cumulative return

```python
cumret = (1 + results).cumprod() - 1
cumret.head(5)
```

#### Annualized return 

```python
total_days = results.shape[0]
total_days
```

```python
cumulative_return = cumret.iloc[-1]
cumulative_return
```

```python
annualized_return = (1 + cumulative_return) ** (251 / total_days) - 1
print(f"Annualized return: \n{round(annualized_return * 100, 2)}%")
```

#### Annualized Volatility

Annualized Volatility = Daily Volatility x $\sqrt{251}$

where:
+ Daily Volatility is the standard deviation of daily return 
+ 251 is the assumed number of trading days in a year

```python
daily_volatility = results.std()
annualized_volatility = daily_volatility * np.sqrt(251)

print(f'Annualized Volatility: \n{round(annualized_volatility * 100, 2)}%')
```

#### Sharpe Ratio
Sharpe Ratio = (Annualized Return - Risk-Free Rate) / Annualized Volatility

where:

+ Annualized return: the annualized return of the portfolio
+ Risk-free rate: the return of a risk free asset. This is typically a small percentage, such as 0.02

```python
risk_free_rate = 0.02
sharpe_ratio = (annualized_return - risk_free_rate) / annualized_volatility

print(f"Sharpe Ratio: \n{round(sharpe_ratio, 2)}")
```

#### Max Drawdown

Max Drawdown = (Trough Value - Peak Value) / Peak Value

Where:

+ Peak Value: The highest portfolio value before the decline
+ Trough Value: The lowest portfolio value during the decline

```python
rolling_max = cumret.cummax()
rolling_max.head()
```

```python
drawdown = (cumret - rolling_max) / rolling_max
max_drawdown = drawdown.min()

print(f"Max Drawdown: \n{round(max_drawdown * 100, 2)}%")
```

1. `cumret.cummax()`: tracks the highest cumulative return up to each point in time
2. `(cumret - rolling_max)/rolling_max`: mesures the percentage decline from the rolling maximum
3. `drawdown.min()`: finds the largest drawdown (most negative value)

#### Sortino Ratio

Sortino Ratio = (Annualized Return - Risk-Free Rate) / Downsize Deviation

Where: 
+ Downside deviation: the standard deviation of negative returns (returns below the risk free rate or a target return)

```python
risk_free_rate = 0.02

excess_negative_return = results[results < risk_free_rate]
excess_negative_return.head()
```

```python
downside_deviation = excess_negative_return.std() * np.sqrt(251)
downside_deviation.head()
```

```python
spotino_ratio = (annualized_return - risk_free_rate) / downside_deviation
print(f'Sortino Ratio: \n {round(spotino_ratio, 2)}')
```

#### Calmar Ratio
The Calmar Ratio measures the risk-adjusted return of an investment by comparing the annulized return the maximum drawdown it is commonly used the evaluate the performance of hedge funds and portfolios

Calmar Ratio = Annualized Return / Maximum Drawdown

```python
calmar_ratio = annualized_return / abs(max_drawdown)
print(f"Calmar Ratio: \n{round(calmar_ratio, 2)}")
```

#### Treynor Ratio
The Treynor Ratio measures the risk adjusted return of an invesment, similar to the sharpe ratio, but it uses systematic risk (beta) instead of total risk (volatility). It evaluates how much excess return is earned per unit of market risk.

Treynor Ratio = (Annualized Return - Risk-free Rate) / Beta

Where:
+ Beta: the portfolio's sensitivity to market movements

```python
risk_free_rate = 0.02

beta = 1.2 # replace with your portfolio's beta

treynor_ratio = (annualized_return - risk_free_rate) /beta

print(f"Treynor Ratio: \n{round(treynor_ratio, 2)}")
```

#### Jensen’s Alpha

The Jensen’s Alpha measures the excess return of a portfolio to compared to its expected return based on the Capital Asset Pricing Model (CAPM). It evaluates how much of a portfolio outperforms (or underperforms) its expected return given its beta and the market's return

Jensen’s Alpha = Portfolio Return - [Risk-Free Rate + Beta x (Market Return - Risk-Free Rate)]

Where:
+ Market Return: the annualized return of the market index

```python
risk_free_rate
beta = 1.2

market_return = 0.1

expected_portfolio_return = risk_free_rate + beta * (market_return - risk_free_rate)

jensens_alpha = annualized_return - expected_portfolio_return

print(f'Jensen’s Alpha: \n {round(jensens_alpha, 2)}')
```
